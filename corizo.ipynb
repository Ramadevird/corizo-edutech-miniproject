# Here we are importing stopwords from nltk package to remove unnecessary words which doesn't give much importance to the sentences

```python
from nltk.corpus import stopwords
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
import nltk
nltk.download('stopwords')
```

```python
stop_words=stopwords.words("english")
```

```python
stop_words
```

```python
port_stemmer=PorterStemmer()#calling the instance.
```

# The process of stemming entails creating morphological variations of a root or base word.Ex- If a phrase contains the word "eating," it changes into the root word "eat."

```python
#Here we are using function for performing stemming to the columns like-title,test
def stemming(columns):
        stemmed_column=re.sub('[^a-zA-Z]',' ',columns)
        stemmed_column=stemmed_column.lower()
        stemmed_column=stemmed_column.split()
        stemmed_column=[port_stemmer.stem(word) for word in stemmed_column if not word in stop_words]
        stemmed_column=' '.join(stemmed_column)
        return stemmed_column
```

```python
df_merge['title']=df_merge['title'].apply(stemming)
df_merge['text']=df_merge['text'].apply(stemming)
```

```python
df_merge#after performing stemming the titleand text doesn't contain any stopwords and it contains only root words as we performed stemming
```

```python
df_merge.drop('date',axis=1,inplace=True)#As there is no time-series analysis in this instance, we are dropping the date column.
```

# tfidfvectorizer is being used to translate the text into values for computer comprehension.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer=TfidfVectorizer()
```

```python
title=df_merge['title'].values
text=df_merge['text'].values
```

```python
tfidf_vectorizer.fit(title)
title_vectorizer=tfidf_vectorizer.transform(title)
```

```python
print(title_vectorizer)
```

```python
tfidf_vectorizer.fit(text)
text_vectorizer=tfidf_vectorizer.transform(text)
```

```python
print(text_vectorizer)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=text_vectorizer
```

```python
y=df_merge["class"].values
```

# here we perform train_test_split.

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=40)
```

# To determine if the news is true or false, we are utilising the logistic regression approach on this dataset.

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg=LogisticRegression()
```

```python
xtrain_pred=log_reg.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
training_data_accuracy=accuracy_score(xtrain_pred,y_train)
```

```python
print("The accuracy of training data is:",training_data_accuracy)
```
